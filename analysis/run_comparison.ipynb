```python
path = "/SWAN_projects/ExtractionTools/data/csv/Many Runs/" # The "Many Runs" folder contains the different menus found in file=/store/data/Run2018C/ZeroBias/NANOAOD/02Apr2020-v2/40000/B3CAAFDF-283A-DF4E-B01F-D819DBEBA1A8.root 
run_1 = "319450"
run_2 = "319840"
run_3 = "320040"
oms_1 = run_1+".csv" # Data taken from the CMSOMS HLT Trigger Report
oms_2 = run_2+".csv" # Data taken from the CMSOMS HLT Trigger Report
oms_3 = run_3+".csv" # Data taken from the CMSOMS HLT Trigger Report

df_oms1 = pd.read_csv(home + path + oms_1).drop(["Sequence","Stream Name"],axis=1)
df_oms2 = pd.read_csv(home + path + oms_2).drop(["Sequence","Stream Name"],axis=1)
df_oms3 = pd.read_csv(home + path + oms_3).drop(["Sequence","Stream Name"],axis=1)
```

```python
# OMS contains version info. Get rid of it. And duplicates.

df_oms1["HLT"] = df_oms1["HLT"].apply(lambda x: x.split("_v")[0] if "_v" in x else x)
df_oms1 = df_oms1.dropna().drop_duplicates()

df_oms2["HLT"] = df_oms2["HLT"].apply(lambda x: x.split("_v")[0] if "_v" in x else x)
df_oms2 = df_oms2.dropna().drop_duplicates()

df_oms3["HLT"] = df_oms3["HLT"].apply(lambda x: x.split("_v")[0] if "_v" in x else x)
df_oms3 = df_oms3.dropna().drop_duplicates()
```

```python
# Create lists for the unique HLT paths found in each menu

l1 = df_oms1["HLT"].unique()
l2 = df_oms2["HLT"].unique()
l3 = df_oms3["HLT"].unique()
```

```python
# Calculate the missing paths for the menus 1 and 2

missing_1_2 = [x for x in l1 if x not in l2]
missing_2_1 = [x for x in l2 if x not in l1]
print(len(missing_1_2), len(missing_2_1))
```

```python
# Calculate the missing paths for the menus 1 and 3. 

missing_1_3 = [x for x in l1 if x not in l3]
missing_3_1 = [x for x in l3 if x not in l1]
print(len(missing_1_3), len(missing_3_1))
```

```python
intersection = [x for x in l1 if x in l3]
```

```python
# Compares the seeds in an HLT path and outputs the missing seeds present in the smaller value
# This is done to investigate if the HTL path with less seeds is contained within the larger HLT pat

for ele in intersection:
    missing = []
    seed1 = df_oms1.loc[df_oms1["HLT"]==ele].Seed.values[0].split(" OR ") 
    seed3 = df_oms3.loc[df_oms3["HLT"]==ele].Seed.values[0].split(" OR ") 
    if len(seed1) > len(seed3):
        missing = [x for x in seed3 if x not in seed1]
    if len(seed3) > len(seed1):
        missing = [x for x in seed1 if x not in seed3]
        
    if missing != []:
        print(ele, missing)
```
